```python
btc_pct = []
btc_pct.append(0)
for i in range(1,len(btc["BTC-USD"])):
  btc_pct.append(((btc["BTC-USD"][i] - btc["BTC-USD"][i-1])/btc["BTC-USD"][i-1]))
btc_pct_change = pd.DataFrame(btc_pct)
```

```python
sum = 0
for i in range(len(ret)):
  sum = sum + (ret["Percent change"][i] - ret_mean)**2
ret_variance = sum/len(ret)
```
